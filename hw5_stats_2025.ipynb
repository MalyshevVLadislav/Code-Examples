Пусть $x_1, \dots, x_n$ &mdash; реализация выборки, $X^{*}_1, X^{*}_2, \dots, X^{*}_n$ &mdash; построенная по ней бутстрепная выборка. С какой вероятностью элемент $x_i$ исходной выборки попадёт в бутстрепную? Посчитайте среднее число уникальных (попарно различных) элементов в бутстрепной выборке, если в исходной выборке все элементы различны? К какой функции от $n$ стремится это количество при $n \rightarrow \infty$?


### 1. Вероятность попадания элемента $x_i$ в бутстрепную выборку

Бутстрепная выборка строится путём $n$ независимых равновероятных выборов с возвращением из исходной выборки $\{x_1, \dots, x_n\}$. Вероятность того, что элемент $x_i$ не будет выбран в одном испытании:  
$P(x_i \text{не выбран}) = 1 - \frac{1}{n}$  
Тогда вероятность того, что $x_i$ не будет выбран ни разу в $n$ испытаниях:  
$P(x_i \text{не попал в бутстреп}) = \left(1 - \frac{1}{n}\right)^n$  
Следовательно, вероятность попадания $x_i$ в бутстрепную выборку хотя бы один раз:  
$P(x_i \text{ попал}) = 1 - \left(1 - \frac{1}{n}\right)^n$  
При $n \to \infty$ получаем:  
$\lim_{n \to \infty} P(x_i \text{ попал}) = 1 - e^{-1} \approx 0.632$  

### 2. Среднее число уникальных элементов в бутстрепной выборке
Пусть $U$ — число уникальных элементов в бутстрепной выборке. Введём индикаторные случайные величины:
$I_i = \begin{cases}
1, & \text{если } x_i \text{ попал в бутстреп}, \\
0, & \text{иначе}
\end{cases}$  
Тогда:  
$U = \sum_{i=1}^n I_i$  
По линейности математического ожидания:  
$\mathbb{E}[U] = \sum_{i=1}^n \mathbb{E}[I_i] = n \cdot \left(1 - \left(1 - \frac{1}{n}\right)^n\right)$  

### 3. Асимптотическое поведение при $n \to \infty$
Используя известный предел:  
$\left(1 - \frac{1}{n}\right)^n \to e^{-1} \quad \text{при} \quad n \to \infty$  
получаем асимптотическое выражение:  
$\mathbb{E}[U] \sim n \left(1 - e^{-1}\right) \approx 0.632n$
