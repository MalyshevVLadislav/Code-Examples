# Проверка гипотез  
### Возможные решения
 - Отклонить $\mathsf{H}_0$ (принять $\mathsf{H}_1$): если $T(x) > 1$. В этом случае делается вывод, что $\theta \notin [-1, 1]$.
 - Не отклонять $\mathsf{H}_0$: если $T(x) \leq 1$. В этом случае недостаточно оснований для отклонения $\mathsf{H}_0$, то есть $\theta \in [-1, 1]$.  

### Вероятность ошибки первого рода
Ошибка первого рода — это вероятность отвергнуть $\mathsf{H}_0$, когда она верна:
$\alpha = \sup_{\theta \in [-1, 1]} \mathsf{P}_{\theta}(T(X) > 1).$

Так как $T(X) \sim \mathcal{N}(\theta, 1)$, то:
$\mathsf{P}_{\theta}(T(X) > 1) = 1 - \Phi(1 - \theta),$
где $\Phi$ — функция стандартного нормального распределения.

Найдем супремум по $\theta \in [-1, 1]$. Рассмотрим поведение $1 - \Phi(1 - \theta)$:
 - При $\theta = 1$: $\mathsf{P}_1(T(X) > 1) = 1 - \Phi(0) = 0.5$.
 - При $\theta \in [-1, 1)$: $1 - \theta \geq 0$, и $\Phi(1 - \theta)$ возрастает при уменьшении $\theta$, значит, $1 - \Phi(1 - \theta)$ убывает.
 - Наибольшее значение достигается при $\theta = 1$.
Таким образом:
$\alpha = \sup_{\theta \in [-1, 1]} \mathsf{P}_{\theta}(T(X) > 1) = \mathsf{P}_1(T(X) > 1) = 0.5.$

**Ответ:** 
### Критерий Вальда
Для проверки гипотезы используем критерий Вальда. В случае двусторонней альтернативы критерий имеет вид:
$S = \left\{ \left|\sqrt{n} \frac{\hat{\theta} - \theta_0}{\hat{\sigma}} \right| > z_{1 - \frac{\alpha}{2}} \right\},$
где:
 - $\hat{\theta}$ — асимптотически нормальная оценка параметра $\theta$,
 - $\hat{\sigma}$ — состоятельная оценка асимптотического стандартного отклонения $\sigma(\theta)$,
 - $z_{1 - \frac{\alpha}{2}}$ — квантиль стандартного нормального распределения уровня $1 - \frac{\alpha}{2}$.
### Асимптотически нормальная оценка параметра $\lambda$
Известно, что для экспоненциального распределения:
$\hat{\lambda} = \frac{1}{\overline{X}}$
является асимптотически нормальной оценкой (АНО) параметра $\lambda$ с асимптотической дисперсией:
$\sigma^2(\lambda) = \lambda^2.$
### Состоятельная оценка дисперсии
Состоятельная оценка асимптотической дисперсии:
$\hat{\sigma}^2 = \hat{\lambda}^2 = \left(\frac{1}{\overline{X}}\right)^2.$

### Статистика критерия Вальда
Подставляем оценки в статистику критерия Вальда:
$W = \sqrt{n} \frac{\hat{\lambda} - \lambda_0}{\hat{\sigma}} = \sqrt{n} \frac{\frac{1}{\overline{X}} - 2}{\frac{1}{\overline{X}}} = \sqrt{n} \left(1 - 2\overline{X}\right).$
Критическая область для уровня значимости $\alpha$:
$S = \left\{ |W| > z_{1 - \frac{\alpha}{2}} \right\}.$
### Эквивалентный доверительный интервал
Асимптотический доверительный интервал для $\lambda$ уровня доверия $1 - \alpha$:
$C = \left( \hat{\lambda} - \frac{z_{1 - \alpha/2} \hat{\sigma}}{\sqrt{n}}, \hat{\lambda} + \frac{z_{1 - \alpha/2} \hat{\sigma}}{\sqrt{n}} \right) = \left( \frac{1}{\overline{X}} - \frac{z_{1 - \alpha/2}}{\overline{X} \sqrt{n}}, \frac{1}{\overline{X}} + \frac{z_{1 - \alpha/2}}{\overline{X} \sqrt{n}} \right).$
### Вывод
Если $|W| > z_{1 - \alpha/2}$, то гипотеза $\mathsf{H}_0$ отклоняется в пользу $\mathsf{H}_1$. В противном случае оснований для отклонения $\mathsf{H}_0$ нет.

На лекции вы узнали про p-value &mdash; это вероятность получить при справедливости $H_0$ такое значение статистики $t = T(x)$ или еще более экстремальное, то есть в случае двустороннего критерия
$$p(x) = \mathsf{P}_0(T(X) \geq|t|) + \mathsf{P}_0(T(X) \leq -|t|)$$
Для расчета можно использовать функции из библиотеки `scipy.stats`.

**2**. Оформите функцию подсчета статистики критерия, p-value и доверительного интервала 

```python
def wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='two_sided'):
    """
    param sample: реализация выборки
    param theta: истинное значение параметра
    param estimation_theta: оценка параметра
    param estimation_sigma: оценка асимптотической дисперсии
        оценки estimation_sigma
    param alternative: вид альтернативной гипотезы,
        может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """

    n = len(sample)
    alpha = 0.05
    statistic = np.sqrt(n) * (estimation_theta - theta) / estimation_sigma
    
    if alternative == 'two_sided':
        p_value = 2 * (1 - sps.norm.cdf(np.abs(statistic)))
    elif alternative == 'less':
        p_value = sps.norm.cdf(statistic)
    elif alternative == 'greater':
        p_value = 1 - sps.norm.cdf(statistic)
    else:
        raise ValueError("alternative must be 'two_sided', 'less' or 'greater'")
    
    z_quantile = sps.norm.ppf(1 - alpha/2)
    margin_of_error = z_quantile * estimation_sigma / np.sqrt(n)
    conf_int = (estimation_theta - margin_of_error, 
                estimation_theta + margin_of_error)
    
    return statistic, p_value, conf_int
```

**3**. Проверьте гипотезу с двусторонней альтернативой.

```python
theta_hat = 1 / np.mean(sample)
sigma_hat = theta_hat
wald_test(sample, 2, theta_hat, sigma_hat, alternative='two_sided')
```

По доверительному интервалу можно судить о том, что истинное значение должно быть существенно выше 2 (так как вся область значений интервала лежит правее 2)

### Задача 3

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем нередко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью сэмплирования по методу Монте-Карло.

Пусть гипотеза $\mathsf{H}_0$ простая, то есть $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. Для ее проверки по конечной выборке $X_1, ..., X_n$ используется критерий $S$. **Опишите схему оценки реального уровня** значимости критерия $S$ методом Монте-Карло.


**Ответ:** 
### Алгоритм оценки

 - Генерация выборок  
  Фиксируем распределение $\mathsf{P}_0$, соответствующее $H_0$, затем генерируем $B$ независимых выборок размера $n$ из $\mathsf{P}_0$:  
$X^{(1)}, ..., X^{(B)} \sim \mathsf{P}_0$

 - Применение критерия:
Для каждой выборки $X^{(b)}$ вычисляем статистику $T^{(b)} = T(X^{(b)})$, затем проверяем попадание в критическую область:
        $I^{(b)} = \mathbb{I}\{T^{(b)} \in S\}$

 - Оценка уровня  
    $\hat{\alpha}_{real} = \frac{1}{B}\sum_{b=1}^B I^{(b)}$  
    С доверительным интервалом:
    $\hat{\alpha}_{real} \pm z_{1-\alpha/2}\sqrt{\frac{\hat{\alpha}_{real}(1-\hat{\alpha}_{real})}{B}}$

### Теоретическое обоснование
По закону больших чисел:
$\hat{\alpha}_{real} \xrightarrow{B\to\infty} \mathsf{P}_0(T(X) \in S) = \alpha_{real}$.

Если гипотеза $\mathsf{H}_0$ сложная, то есть $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, то можно повторить такую процедуру несколько раз для разных $\mathsf{P} \in \mathscr{P}_0$.

Рассмотрим альтернативную гипотезу $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. **Опишите схему оценки мощности** критерия $S$ методом Монте-Карло.

**Ответ:**  
### Алгоритм оценки

 - Выбор параметров  
Выбираем конкретное распределение $\mathsf{P}^* \in \mathscr{P}_1$, затем фиксируем размер выборки $n$ и количество повторений $B$.
 - Генерация выборок
    $X^{(1)}, \dots, X^{(B)} \sim \mathsf{P}^*$
 - Применение критерия  
Для каждой выборки $X^{(b)}$ вычисляем статистику $T^{(b)} = T(X^{(b)})$, затем проверяем попадание в критическую область:  
$I^{(b)} = \mathbb{I}\{T^{(b)} \in S\}$

 - Оценка мощности  
$\hat{\beta}(\mathsf{P}^*) = \frac{1}{B}\sum_{b=1}^B I^{(b)}$  
С доверительным интервалом:  
$\hat{\beta}(\mathsf{P}^*) \pm z_{1-\alpha/2}\sqrt{\frac{\hat{\beta}(\mathsf{P}^*)(1-\hat{\beta}(\mathsf{P}^*))}{B}}$

#### Особенности для сложной гипотезы
Для получения полной картины:
 - Повторить процедуру для нескольких $\mathsf{P}^* \in \mathscr{P}_1$
 - Построить график зависимости $\hat{\beta}(\mathsf{P}^*)$ от параметров распределения
 - Для $H_0$ оценить реальный уровень значимости на "граничных" распределениях $\mathsf{P} \in \mathscr{P}_0$

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените реальный уровень значимости** критерия для $ n = 50$.

```python
def wald_test_poisson(sample, theta0=1, alpha=0.05):
    n = len(sample)
    theta_hat = np.mean(sample)
    se = np.sqrt(theta_hat / n)
    W = (theta_hat - theta0) / se
    p_value = 2 * (1 - sps.norm.cdf(np.abs(W)))
    return W, p_value

def estimate_true_level(n=50, theta0=1, B=10000, alpha=0.05):
    rejections = 0
    
    for _ in range(B):
        sample = sps.poisson.rvs(mu=theta0, size=n)
        
        W, p_value = wald_test_poisson(sample, theta0)
        
        if p_value < alpha:
            rejections += 1
    
    alpha_hat = rejections / B
    se_hat = np.sqrt(alpha_hat * (1 - alpha_hat) / B)
    
    return alpha_hat, se_hat

n = 50
B = 10000
alpha = 0.05

alpha_hat, se_hat = estimate_true_level(n=n, B=B)
print(f"Оценка реального уровня значимости: {alpha_hat:.4f} ± {1.96*se_hat:.4f} (95% ДИ)")
print(f"Относительная ошибка: {100*(alpha_hat - alpha)/alpha:.1f}%")
```

**Вывод:** для n = 50 реальный уровень значимости (4.8%) близок к ожидаемым 5%, есть небольшая относительная ошибка в 0.4%. Вероятно, что при уменьшении n полученные результаты могут быть хуже из-за асимптотичности метода Монте-Карло и наоборот, при увеличении n полученные результаты должны становиться только лучше.

**2.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам при $\theta=0.2$ и при $\theta=2$.

```python
def wald_test(sample, theta0=1, alternative='two-sided'):
    n = len(sample)
    theta_hat = np.mean(sample)
    se = np.sqrt(theta_hat / n)
    W = (theta_hat - theta0) / se
    
    if alternative == 'two-sided':
        p_value = 2 * (1 - sps.norm.cdf(abs(W)))
    elif alternative == 'less':
        p_value = sps.norm.cdf(W)
    elif alternative == 'greater':
        p_value = 1 - sps.norm.cdf(W)
    return W, p_value

def estimate_power(n, true_theta, B=10000, alpha=0.05):
    results = {}
    
    for alt in ['two-sided', 'less', 'greater']:
        rejections = 0
        
        for _ in range(B):
            sample = sps.poisson.rvs(mu=true_theta, size=n)
            _, p_value = wald_test(sample, theta0=1, alternative=alt)
            if p_value < alpha:
                rejections += 1
                
        power = rejections / B
        results[alt] = power
    
    return results

n = 50
B = 10000
theta_values = [0.2, 2]

for theta in theta_values:
    print(f"\nθ = {theta}:")
    powers = estimate_power(n=n, true_theta=theta, B=B)
    
    for alt, power in powers.items():
        print(f"{alt}: {power}")
```

**Вывод:** видим, что если альтернатива меньше 1 (случай $\theta = 0.2$), правосторонняя альтернатива вообще не имеет смысла, в то время как левосторонняя и двусторонняя альтернатива дают одинаковые результаты в пользу отвержения нулевой гипотезы (на самом деле левосторонняя должна давать более лучший результат, но здесь за счет округления результаты кажутся эквивалентными). Для альтернативы больше 1 (случай $\theta = 2$) все наоборот. Левосторонняя альтернатива не дает результата, а вот правосторонняя и двусторонняя отвергают нулевую гипотезу (здесь как раз и видим различие - специализированная альтернатива (в данном случае правосторонняя) показывает себя лучше, чем более универсальная двусторонняя. Тем не менее, результат двусторонней альтернативы все еще отвергает нулевую гипотезу, а значит она также применима в данном случае).

2. Пусть  $X = ( X_1, ... X_n)$ &mdash; выборка из неизвестного распределения $\mathsf{P}$ с конечным вторым моментом. Постройте асимптотический критерий для проверки гипотез $\mathsf{H}_0 \colon \mathsf{E} X = 0 $   vs.  $\mathsf{H}_1 \colon \mathsf{E} X > 0$ уровня значимости $\alpha$. Покажите, что при $n \to \infty$ этот критерий совпадает с критерием из первого пункта.

**Решение:**
## Одновыборочный t-критерий для среднего нормального распределения
### Статистика критерия
Используем t-статистику:  
$T(X) = \sqrt{n - 1} \frac{\bar{X}}{s}$  
где:  
 - $\bar{X} = \frac{1}{n}\sum_{i=1}^n X_i$ — выборочное среднее  
 - $s = \sqrt{\frac{1}{n}\sum_{i=1}^n (X_i - \bar{X})^2}$ — несмещенная оценка стандартного отклонения.  
### Распределение статистики  
При $H_0$ (т.е. $a = 0$):  
$T(X) \sim t_{n-1},$  
где $t_{n-1}$ — распределение Стьюдента с $n-1$ степенями свободы.
### Критическая область
Критерий уровня $\alpha$:
$S = \{ T(X) > t_{n-1, 1-\alpha} \},$  
где $t_{n-1, 1-\alpha}$ — $(1-\alpha)$-квантиль распределения $t_{n-1}$.

### P-value
Для правосторонней альтернативы:
$p\text{-value} = \mathbb{P}(t_{n-1} > T_{\text{obs}})$
где $T_{\text{obs}}$ — наблюдаемое значение статистики.
### Связь с классической t-статистикой  
Заметим, что:  
$\sqrt{n-1}\frac{\bar{X}}{s} = \sqrt{n}\frac{\bar{X}}{s'} \cdot \sqrt{\frac{n-1}{n}}$  
где $s'$ — исправленное стандартное отклонение. При больших $n$ разница несущественна.

### Асимптотическая статистика
По ЦПТ и теореме Слуцкого:
$T_n = \sqrt{n-1} \frac{\bar{X}}{s_n} = \sqrt{\frac{n-1}{n}} \cdot \sqrt{n} \frac{\bar{X}}{s_n} \xrightarrow{d} \mathcal{N}(0,1)$  
так как:  
 - $\sqrt{n}\bar{X} \xrightarrow{d} \mathcal{N}(0,\sigma^2)$
 - $s_n \xrightarrow{p} \sigma$
 - $\sqrt{\frac{n-1}{n}} \to 1$
### Критическая область  
Асимптотический критерий уровня $\alpha$:  
$S = \{ T_n > z_{1-\alpha} \},$  
где $z_{1-\alpha}$ — $(1-\alpha)$ - квантиль стандартного нормального распределения.
### Связь с t-критерием  
При $n \to \infty$:
 - $t_{n-1}$-распределение сходится к $\mathcal{N}(0,1)$
 - Критические значения: $t_{n-1,1-\alpha} \to z_{1-\alpha}$
 - Таким образом, t-критерий становится эквивалентным асимптотическому критерию
