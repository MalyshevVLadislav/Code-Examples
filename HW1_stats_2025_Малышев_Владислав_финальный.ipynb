### Задача 1

Пусть $X_1,...,X_n$ выборка из из распределения $U(0, \theta)$. На лекции во время практики вы уже визуализировали свойство несмещенности для следующих оценок параметра $\theta$: $2\overline{X}$, $X_{(n)}$, $\frac{n+1}{n}X_{(n)}$. Теперь наша задача убедиться, согласуется ли практика с теорией. Для решения этой задачи воспользуйтесь утверждениями с лекции.

*Указание.* Найдите функцию распределения случайной величины $X_{(n)}$, посчитав вероятность того, что максимальное значение среди элементов выборки не превосходит числа $x$. Используйте при этом определение выборки. После чего получите плотность $X_{(n)}$.


Для начала найдем функцию распределения случайной величины $X_{(n)}$:  
$F_{X_{(n)}}(x) = P(X_{(n)} \leq x) = P(X_1 \leq x,..., X_{(n)} \leq x),$  
а так как все $X_(i)$ независимы и одинаково распределены, то:  
$F_{X_{(n)}}(x) = \prod\limits_{i=1}^n P(X_i \leq x) = \left( F_X(x) \right)^n,$  
где $F_X(x)$ - функция распределения $X_i \sim U(0, \theta),$ которая выражается как:  
$F_X(x) = 
\begin{cases}
0,      & x \leq 0, \\
\frac{x}{\theta}, & 0 < x < \theta, \\
1,      & x \geq \theta.
\end{cases}$  
Тогда имеем формулу:  
$F_{X_{(n)}}(x) =
\begin{cases}
0,      & x \leq 0, \\
(\frac{x}{\theta})^n, & 0 < x < \theta, \\
1,      & x \geq \theta.
\end{cases}$  
Найдем плотность распределения $X_{(n)}$:  
$f_{X_{(n)}}(x) = \frac{d}{dx} F_{X_{(n)}}(x) = 
\begin{cases}
\frac{n x^{n-1}}{\theta^n}, & 0 < x < \theta, \\
0, & \text{иначе}.
\end{cases}$  
Теперь можем начать проверять оценки на несмещенность:  
1) $2\overline{X}$:  
   $E[2\overline{X}] = 2E[\frac{1}{n}\sum\limits_{i=1}^{n}X_{i}] = 2 * \frac{1}{n} \sum\limits_{i=1}^{n}E[X_{i}] = 2 * \frac{1}{n} * n * \frac{\theta}{2} = \theta \Rightarrow$ оценка $2\overline{X}$ является несмещенной оценкой $\theta$.
2) $X_{(n)}$:  
   $E[X_{(n)}] = \int\limits_{-\infty}^{\infty} x f_{X_{(n)}}(x) \, dx = \int\limits_{0}^{\theta} x \cdot \frac{n x^{n-1}}{\theta^n} \, dx = \frac{n}{\theta^n} \int\limits_{0}^{\theta} x^n \, dx = \frac{n}{\theta^n} * \frac{\theta^{n + 1}}{n + 1} = \frac{n}{n + 1}\theta \Rightarrow$ оценка $X_{(n)}$ неявляется несмещенной оценкой $\theta$ (на самом деле является ассимптотически несмещенной, нам потом это на лекции расскажут вроде как).
3) $\frac{n + 1}{n}X_{(n)}$:  
   $E[\frac{n + 1}{n}X_{(n)}] = \frac{n + 1}{n}E[X_{(n)}] = \frac{n + 1}{n} \frac{n}{n + 1}\theta = \theta \Rightarrow$ оценка $\frac{n + 1}{n}X_{(n)}$ является несмещенной оценкой $\theta$.


### Задача 2


Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно.
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.


2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.


3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

1) Докажем, что $S^2 = \overline{X^2}-\overline{X}^2$:  
   $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2 = \frac{1}{n}\sum\limits_{i=1}^n(X_i^2 - 2X_i\overline{X} + \overline{X}^2) = \frac{1}{n}\sum\limits_{i=1}^nX_i^2 - \frac{2}{n}\overline{X}\sum\limits_{i=1}^nX_i + \frac{1}{n}\sum\limits_{i=1}^n\overline{X}^2 = \frac{1}{n} * n\overline{X^2} - \frac{2}{n}\overline{X} * n\overline{X} + \frac{1}{n} * n\overline{X}^2 = \overline{X^2} - \overline{X}^2 \Rightarrow$ выражение верно.
2) По определению статистика $S^2$ будет являться состоятельной оценкой дисперсии $\sigma^2$, если $S^2 \xrightarrow{P} \sigma^2 \quad \text{при} \quad n \to \infty$.
   Из ЗБЧ имеем $\overline{X} \xrightarrow{P} E[X_1] = \mu$ и $\overline{X^2} \xrightarrow{P} E[X_1^2] = \sigma^2 + \mu^2$, а по теореме Манна-Вальда $\overline{X}^2 \xrightarrow{P} E[X_1] = \mu^2$, откуда:  
$S^2 = \overline{X^2} - \overline{X}^2 \xrightarrow{P} (\sigma^2 + \mu^2) - \mu^2 = \sigma^2 \Rightarrow S^2$ - состоятельная оценка $\sigma^2$.
3) По определению статистика $S^2$ будет являться несмещенной оценкой $\sigma^2$ если $E[S^2] = \sigma^2$. Тогда найдем, чему равно $E[S^2]$:
   $E[S^2] = E[\overline{X^2}-\overline{X}^2] = E[\overline{X^2}] - E[\overline{X}^2] = \sigma^2 + \mu^2 - (Var(\overline{X}) + (E[\overline{X}])^2) = \sigma^2 + \mu^2 - (Var(\frac{1}{n}\sum\limits_{i=1}^nX_i) + \mu^2) = \sigma^2 - \frac{1}{n^2}Var(\sum\limits_{i=1}^nX_i)   
= \text{(так как случайные величины независимы, то дисперсия суммы равна сумме дисперсий)} = \sigma^2 - \frac{1}{n^2}\sum\limits_{i=1}^nVar(X_i) = \sigma^2 - \frac{1}{n^2} * n\sigma^2 = \frac{n - 1}{n}\sigma^2 \neq \sigma^2 \Rightarrow$ оценка $S^2 = \overline{X^2}-\overline{X}^2$ является смещенной.

### Задача 3


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность, повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок (первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return np.var(X, axis=1)

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return np.var(X, axis=1) * len(X[0]) / (len(X[0]) - 1)
    
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

Напишите функцию построения таких графиков в общем виде: на вход функция может получать любые распределения (`distributions`) и любые оценки (`estimators`). 
Средние значения по экспериментам для каждой оценки нужно обозначить звездочками.

```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))
    plt.figure(figsize=(20, len(estimators)))
    for estimator, name, color, y in estimators:
        D = estimator(X)
        plt.scatter(D, np.zeros(sample_count) + y, alpha=0.1,
                        s=100, color=color, label=name)
        plt.scatter(D.mean(), y, marker='*', s=300,
                        color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend(loc='upper right')
        plt.xlim((0.8, 1.2))
        plt.ylim((-0.1, 0.2 * len(estimators)))
```

Постройте три графика для разных размеров выборок (10, 100, 500).


```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

for sample_size in sample_size_list:
    est_plot(sps.norm(loc=0), estimators, sample_size, sample_count)
```

```python
def var_plot(distribution, estimators, n_grid, sample_count):
    '''
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    n_grid -- массив значений размера выборки
    sample_count -- количество генерируемых выборок
    '''
    n_grid = n_grid.astype(int)
    X = distribution.rvs(size = (sample_count, len(n_grid)))
    plt.figure(figsize=(15, 3*len(estimators)))
    plt.hlines(1, 0, len(n_grid), color='r', label=r'$\sigma^2$')
    for estimator, name, color, y in estimators:
        D = np.array([np.mean(estimator(X[0:len(n_grid), 0:(n_grid[i]-1)])) for i in range(len(n_grid))])
        plt.plot(n_grid, D, color= color, label= name)
        plt.xlabel('размер выборки')
        plt.xlim(10, 100)
        plt.legend()
        plt.title('зависимость среднего оценок от размера выборки')
```

```python
n_grid = np.linspace(10, 300, 100)
var_plot(sps.norm, estimators, n_grid, sample_count)
```

В данной задаче мы предполагаем, что распределение имеет экспоненциальный характер, оценим параметр этого распределения.

Для этого выведем ОМП для параметра экспоненциального распределения:

Функция правдоподобия имеет вид:
$L(\lambda) = \prod\limits_{i=1}^n f(X_i|\lambda) = \lambda^n \exp\left(-\lambda\sum\limits_{i=1}^n X_i\right)$
Теперь возьмем от нее натуральный логарифм, продифференцируем и приравняем производную к 0:
$\frac{\partial \ln(L)}{\partial \lambda} = \frac{n}{\lambda} - \sum_{i=1}^n X_i = 0.$
Таким образом получим $\lambda_{\text{ОМП}} = \frac{n}{\sum_{i=1}^n X_i} = \frac{1}{\overline{X}}$.  
Теперь проверка на максимум:  
Вторая производная:
$\frac{\partial^2 \ln(L)}{\partial \lambda^2} = -\frac{n}{\lambda^2} < 0,$ что подтверждает наличие максимума.

```python
x = np.arange(0, 500, 20)
l_1 = 1 / X.mean()
l_2 = (len(X) - 1) / len(X) * l_1
plt.hist(X, bins=20, density = True, alpha=0.5)
plt.plot(x, l_1 * np.exp(-l_1 * x), 'r-', label=r'$Exp(\lambda_{opt})$')
plt.plot(x, l_2 * np.exp(-l_2 * x), 'r-', label=r'$Exp(\lambda)$')
plt.xlabel("Число пятен")
plt.ylabel("Частота")
plt.title(f'Гистограмма числа темных пятен на солнце в день')
plt.legend(loc='upper right')
plt.show()
```

**Вывод:** С помощью полученной оценки нам удается доказать, что распределение исследуемой величины имеет именно экспоненциальный характер. За счет большой выборки не видно того, что на самом деле оценка $\lambda = \frac{1}{\overline{X}}$ не является несмещенной (она асимптотически несмещенная). Как видим, после нанесения функции плотности для несмещенной оценки, картина не меняется, что позволяет в данной задаче пользоваться исходно полученной оценкой.

### Задача 5

Рассмотрим модель линейной регрессии

$$Y = X \theta + \epsilon,$$

где $Y \in \mathbb{R}^n$ - отклик, $X \in \mathbb{R}^{n \times d}$, $\theta \in \mathbb{R}^d$, $\epsilon \in \mathbb{R}^n$ - шум, $n > d$, $rk X = d$.

1. Опишите, какие величины в модели являются
  - случайными
  - неизвестными
2. Покажите, что если $E \epsilon = 0$, то МНК-оценка $\widehat{\theta}$ параметра $\theta$

  $$\widehat{\theta} = (X^TX)^{-1}X^T Y$$

  является несмещенной оценкой параметра $\theta$. Будет ли выполнено свойство несмещенности для Ridge-регрессии?

  *Указание:* используйте свойство линейности математического ожидания в матричном виде.
  
3. Покажите, что, если дополнительно матрица ковариаций $D \epsilon = \sigma^2 I_n$, дисперсия МНК-оценки $D \widehat{\theta} = \sigma^2 (X^TX)^{-1}$. Что показывает данный результат в случае мультиколлинеарности признаков?
  *Указание:* используйте свойство дисперсии в матричном виде.


**Напоминание**

1. Матричное свойство математического ожидания:

  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$E A \psi = A E \psi$$
2. Матричное свойство дисперсии:
  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$D A \psi = A (D \psi) A^T$$

### 1. Случайные и неизвестные величины в модели

В модели линейной регрессии:  
$Y = X \theta + \epsilon,$  
1) Случайные величины:  
  - Вектор откликов $Y$ (зависит от случайного шума $\epsilon$),  
  - Сам вектор шума $\epsilon$.  
2) Неизвестные величины:  
  - Вектор параметров $\theta$,  
  - Дисперсия шума $\sigma^2$ (если $D[\epsilon] = \sigma^2 I_n$) (хотя конечно нас это редко интересует, это больше интересно именно в случаях когда мы и исследуем этот шум).  
### 2. Несмещенность МНК-оценки

МНК-оценка:  
$\widehat{\theta} = (X^T X)^{-1} X^T Y.$  
Подставим $Y = X \theta + \epsilon$:  
$\widehat{\theta} = (X^T X)^{-1} X^T (X \theta + \epsilon) = \theta + (X^T X)^{-1} X^T \epsilon.$  
Математическое ожидание (с учетом $E \epsilon = 0$ и линейности $E$):  
$E \widehat{\theta} = \theta + (X^T X)^{-1} X^T \cdot 0 = \theta.$  
Вывод: $\widehat{\theta}$ — несмещенная оценка $\theta$.  
Для Ridge-регрессии:  
Оценка $\widehat{\theta}_{\text{ridge}} = (X^T X + \lambda I)^{-1} X^T Y$ смещена, так как:  
$E \widehat{\theta}_{\text{ridge}} = (X^T X + \lambda I)^{-1} X^T X \theta \neq \theta \quad (\text{при } \lambda > 0).$  
### 3. Дисперсия МНК-оценки

При $D \epsilon = \sigma^2 I_n$:  
$D \widehat{\theta} = D \left[(X^T X)^{-1} X^T Y\right] = (X^T X)^{-1} X^T \cdot D Y \cdot X (X^T X)^{-1}.$  
Так как $D Y = D \epsilon = \sigma^2 I_n$:  
$D \widehat{\theta} = \sigma^2 (X^T X)^{-1} X^T X (X^T X)^{-1} = \sigma^2 (X^T X)^{-1}.$  

Интерпретация при мультиколлинеарности:  
Если столбцы \(X\) почти линейно зависимы, матрица \(X^T X\) близка к вырожденной, её обратная \((X^T X)^{-1}\) имеет большие собственные значения. Это приводит к росту дисперсии оценки $\widehat{\theta}$.

  - Учитывая ответы в задаче 5, опишите, как должны генерироваться выборки для визуализации. Какие величины в формулах из п.1 должны быть зафиксированы? Какие должны меняться от эксперимента к эксперименту? Каким образом?
  - Задайте сами истинное значение $\theta$ и значения признаков $X \in \mathcal{R}^{n \times d}$. Учитывайте, что для выполнения условия теоремы с лекции $n > d$ и $rk X = d$. Для Ridge-регрессии возьмите два значения $\lambda$. Сгенерируйте зашумленные значения $Y$, с помощью которых можно получить выборки для каждого вида оценки. Визуализируйте их аналогично задаче 3. Распределение шума $\epsilon$ можно считать нормальным. 
  
  **Внимание:** вы должны получить $m$ реализаций $Y$, по которым будут посчитаны $m$ реализаций оценок каждого вида.

### 1. Формулы оценок

МНК-оценка:  
$\widehat{\theta}_{\text{МНК}} = (X^T X)^{-1} X^T Y$

Ridge-оценка (с параметром регуляризации $\lambda > 0$):  
$\widehat{\theta}_{\text{Ridge}} = (X^T X + \lambda I)^{-1} X^T Y$

### Генерация данных:
1. Фиксированные величины:
   - Матрица признаков $X \in \mathbb{R}^{n \times d}$ ($n > d$, $\text{rk}(X) = d$).
   - Истинный параметр $\theta \in \mathbb{R}^d$ (задаётся нами согласно второй точке произвольно).
   - Параметр $\lambda$ для Ridge (аналогично задаем мы, возьмём $\lambda_1 = 0.1$, $\lambda_2 = 1$).

2. Случайные величины:
   - Шум $\epsilon \sim \mathcal{N}(0, \sigma^2 I_n)$ (например, $\sigma = 0.2$).
   - Вектор $Y = X \theta + \epsilon$ генерируется заново для каждой из $m$ реализаций.

```python
np.random.seed(42)
n = 100  
d = 1    
m = 100 
sigma = 0.5  
lambdas = [1, 10]  #сразу возьмем значение побольше чтобы результат было лучше видно

X = np.random.randn(n, d)
X[:, 0] = 1  # для полноты эксперимента добавим константу
theta_true = np.array([1.5])  #например такие, так как этот параметр мы ставим сами

theta_mnk = np.zeros((m, d))
theta_ridge = {lam: np.zeros((m, d)) for lam in lambdas}

for i in range(m):
    epsilon = np.random.normal(0, sigma, n)
    Y = X @ theta_true + epsilon
    
    mnk = LinearRegression(fit_intercept=False).fit(X, Y)
    theta_mnk[i] = mnk.coef_
    
    for lam in lambdas:
        ridge = Ridge(alpha=lam, fit_intercept=False).fit(X, Y)
        theta_ridge[lam][i] = ridge.coef_

samples = (0, 1, 10)  

for s in samples:
    if s == 0:
        theta = theta_mnk
        name = r'$\widehat{\theta}_{МНК}$'
    else:
        theta = theta_ridge[s]
        name = r'$\widehat{\theta}_{Ridge}, \lambda = $' +  f'{s}'
    plt.plot(theta, label=name)
plt.plot([1.5 for i in range(m)], label=r'$\theta_{ист}$')    
plt.xlabel('номер реализации')
plt.ylabel(r'значение оценки $\theta$')
plt.xlim(10, 100)
plt.legend(loc='upper right')
plt.title('зависимость среднего оценок для разных реализаций Y')
plt.show()
```
